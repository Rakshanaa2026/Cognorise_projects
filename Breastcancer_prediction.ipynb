```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/breastcancer_data.csv')
```

```python
df = df.drop(columns=['id','Unnamed: 32'],axis=1)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df['diagnosis'] = encoder.fit_transform(df['diagnosis'])
```

```python
unique_values = df['diagnosis'].unique()
unique_values
```

```python
correlation = df.corr()
correlation
```

```python
df.groupby('diagnosis').mean()
```

```python
X = df.drop(columns='diagnosis',axis=1)
```

```python
Y = df['diagnosis']
```

```python
Y.tail()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(xtrain.shape,xtest.shape,X.shape)
print(ytrain.shape,ytest.shape,Y.shape)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
xtrain_prediction = model.predict(xtrain)
train_accuracy = accuracy_score(ytrain,xtrain_prediction)
train_accuracy
```

```python
xtest_prediction = model.predict(xtest)
test_accuracy = accuracy_score(ytest,xtest_prediction)
print(test_accuracy)
```
